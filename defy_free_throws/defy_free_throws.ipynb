```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns; sns.set(style='white')
import numpy as np
import pandas as pd

from nba_api.stats.endpoints import LeagueDashTeamStats
from nba_api.stats.static import teams
from nba_api.stats.library import parameters as pp
```

```python
keep_cols = ['TEAM_ID', 'TEAM_NAME', 'GP', 'W', 'L', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'PTS',
             'FTM_RANK', 'FTA_RANK', 'PTS_RANK']
```

```python
df_pre = LeagueDashTeamStats(
    per_mode_detailed=pp.PerModeDetailed.per_game,
    season_segment_nullable=pp.SeasonSegment.pre_all_star
).get_data_frames()[0][keep_cols]
df_pre.head(2)
```

```python
df_post = LeagueDashTeamStats(
    per_mode_detailed=pp.PerModeDetailed.per_game,
    season_segment_nullable=pp.SeasonSegment.post_all_star
).get_data_frames()[0][keep_cols]
df_post.head(2)
```

```python
df_all = LeagueDashTeamStats(
    per_mode_detailed=pp.PerModeDetailed.per_game,
#     season_segment_nullable=Nonepp.SeasonSegment.
).get_data_frames()[0][keep_cols]
df_all.head(2)
```

```python
df_post.sort_values('FTM_RANK')
```

```python
for d in [df_pre, df_post, df_all]:
    d['PTS_MINUS_FTM'] = d.eval('PTS - FTM')
    d['PTS_MINUS_FTM_RANK'] = d.PTS_MINUS_FTM.rank(method='first').astype(int)
df_post.sort_values('PTS_MINUS_FTM_RANK')
```

```python
ls = LeagueStandings().get_data_frames()[0]
team2div = ls.set_index('TeamID').Division.to_dict()
team2conf = ls.set_index('TeamID').Conference.to_dict()

ls.head(2)
```

```python
team2abr = pd.DataFrame(teams.get_teams()).set_index('id').abbreviation.to_dict()
```

```python
for d in [df_pre, df_post, df_all]:
    d['division'] = d.TEAM_ID.map(team2div)
    d['conference'] = d.TEAM_ID.map(team2conf).map({'East': '', 'West': ''})
    d['abr'] = d.TEAM_ID.map(team2abr)
    d['WIN_PCT'] = d.eval('W / (W + L) * 100')#.rank()
    d['ft_rate'] = d.eval('FTM / PTS * 100')
```

```python
df_cmp = pd.concat([df_pre.set_index('TEAM_ID').ft_rate.to_frame('before'),
                    df_post.set_index('TEAM_ID').ft_rate.to_frame('after')], axis=1).reset_index()
df_cmp['dif'] = df_cmp.eval('after - before')
df_cmp['dif_abs'] = df_cmp.eval('after - before').abs()
df_cmp.sort_values('dif_abs', ascending=False, inplace=True)
df_cmp['abr'] = df_cmp.TEAM_ID.map(team2abr)
df_cmp['conference'] = df_cmp.TEAM_ID.map(team2conf).map({'East': '', 'West': ''})
df_cmp.head(10)
```

```python
df_all.sort_values('WIN_PCT').head(10)
```

```python
ch = alt.Chart(df_cmp, height=600, width=600, title=' ')
min_ = min(df_cmp.before.min(), df_cmp.after.min())
max_ = max(df_cmp.before.max(), df_cmp.after.max())

points = (
    ch
    .mark_point()
    .encode(x=alt.X("before", 
                    scale=alt.Scale(domain=[min_, max_])
                   ),
            y=alt.Y("after", 
                    scale=alt.Scale(domain=[min_, max_])
                   ),
            color=alt.Color("conference", title=''),
#             size=alt.Size('WIN_PCT', title=' ', scale=alt.Scale(domainMin=df_all.WIN_PCT.min(),
#                                                                             domainMax=df_all.WIN_PCT.max())),
           )
)

labels = (
    ch
    .mark_text(angle=360,
        align='left',
        baseline='middle',
        dx=7, 
        size=15
    )
    .encode(x=alt.X("before", title='   ',
#                     scale=alt.Scale(domain=[df_all.PTS.min(), df_all.PTS.max()])
                   ),
            y=alt.Y("after", title='   ',
#                     scale=alt.Scale(domain=[df_all.PTS_MINUS_FTM.min(), df_all.PTS_MINUS_FTM.max()])
                   ),
            text='abr', color=alt.Color("conference") #, scale=color_scale)
           )
)

line = pd.DataFrame({
    'before': [min_, max_],
    'after': [min_, max_],
})

line_plot = alt.Chart(line).mark_line(color= 'red').encode(
    x= 'before',
    y= 'after',
)
points + labels + line_plot
```

```python
df_tc = pd.read_csv('scoring_leaders.csv').iloc[-25:].rename(columns={'Pos[c]': 'pos'})
df_tc['x'] = (
    df_tc.Season 
)
df_tc['name'] = (
    #    + ' ' 
    df_tc.Player.map(lambda s: " ".join(s.replace('*', ' ').replace('^', ' ').replace('', ' ').split()[:2]))
)
df_tc['ft_share'] = (df_tc['Free throws\nmade'] / df_tc['Total\npoints']) * 100
df_tc
```

```python
ch = alt.Chart(df_tc, height=600, width=800)

points = (
    ch
    .mark_point()
    .encode(x=alt.X("x", title=''
#                     scale=alt.Scale(domain=[df_all.PTS.min(), df_all.PTS.max()])
                   ),
            y=alt.Y("ft_share", title=' ,    '
#                     scale=alt.Scale(domain=[df_all.PTS_MINUS_FTM.min(), df_all.PTS_MINUS_FTM.max()])
                   ),
            color=alt.Color("pos", title=''),
            size=alt.Size('Points pergame', title='    ', 
                          scale=alt.Scale(domainMin=df_tc['Points pergame'].min() - 3,       
                                          domainMax=df_tc['Points pergame'].max() + 4)
                         ),
           )
)

labels = (
    ch
    .mark_text(angle=270,
        align='right',
        baseline='middle',
        dx=-10,
               dy=0,
        size=15
    )
    .encode(x=alt.X("x"),
            y=alt.Y("ft_share"),
            text='name', color=alt.Color("pos") #, scale=color_scale)
           )
)
points  + labels
```
